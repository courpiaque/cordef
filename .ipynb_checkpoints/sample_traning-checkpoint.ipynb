```python
converted_files_train_path = '/content/deft_corpus/data/converted_files_train'
converted_files_dev_path = '/content/deft_corpus/data/converted_files_dev'

files_train = os.listdir(converted_files_train_path)
files_train = [os.path.join(converted_files_train_path, f) for f in files_train]

files_dev = os.listdir(converted_files_dev_path)
files_dev = [os.path.join(converted_files_dev_path, f) for f in files_dev]
```

```python
df_train = []
df_dev = []

for f in files_train:
  with open(f, 'r', encoding='utf-8') as text_file:
    content = text_file.readlines()
    content = [line.strip() for line in content if line != '']

    for line in content:
      sentence, label = line.split('\t')
      df_train.append((sentence[2:-1], label[3:-1]))

for f in files_dev:
  with open(f, 'r', encoding='utf-8') as text_file:
    content = text_file.readlines()
    content = [line.strip() for line in content if line != '']

    for line in content:
      sentence, label = line.split('\t')
      df_dev.append((sentence[2:-1], label[3:-1]))
```

```python
len(df_train), len(df_dev)
```

```python
df_train = pd.DataFrame(df_train, columns=['Sentence', 'Label'])
df_train
```

```python
df_dev = pd.DataFrame(df_dev, columns=['Sentence', 'Label'])
df_dev
```

# SpaCy

## Konwersja do formatu spaCy

```python
train_data_tokens = []
train_data_labels = []
train_data_sentences = []
for index, row in df_train.iterrows():
  sentence = f'Sentence {index+1}'
  temp_tokens = []
  temp_labels = []
  temp_sentences = []
  for token in row['Sentence'].split():
    temp_tokens.append(token)
    temp_sentences.append(sentence)
  train_data_tokens.append(temp_tokens)
  train_data_sentences.append(temp_sentences)
  for label in row['Label'].split():
    temp_labels.append(label)
  train_data_labels.append(temp_labels)

df_train['token_list'] = train_data_tokens
df_train['label_list'] = train_data_labels
df_train['sentence_list'] = train_data_sentences
```

```python
dev_data_tokens = []
dev_data_labels = []
dev_data_sentences = []
for index, row in df_dev.iterrows():
  sentence = f'Sentence {index+1}'
  temp_tokens = []
  temp_labels = []
  temp_sentences = []
  for token in row['Sentence'].split():
    temp_tokens.append(token)
    temp_sentences.append(sentence)
  dev_data_tokens.append(temp_tokens)
  dev_data_sentences.append(temp_sentences)
  for label in row['Label'].split():
    temp_labels.append(label)
  dev_data_labels.append(temp_labels)

df_dev['token_list'] = dev_data_tokens
df_dev['label_list'] = dev_data_labels
df_dev['sentence_list'] = dev_data_sentences
```

```python
train_output = []
train_output_hg = [('Sentence #', 'Word', 'Tag')]
for sentence_list, token_list, label_list in zip(train_data_sentences, train_data_tokens, train_data_labels):
  for sentence, token, label in zip(sentence_list, token_list, label_list):
    line = f'{token}\t{label}'
    train_output.append(line)
    line = (sentence, token, label)
    train_output_hg.append(line)

with open('train.iob', 'w', encoding='utf-8') as iob_file:
  iob_file.write('\n'.join(train_output))
```

```python
dev_output = []
dev_output_hg = [('Sentence #', 'Word', 'Tag')]
for sentence_list, token_list, label_list in zip(dev_data_sentences, dev_data_tokens, dev_data_labels):
  for sentence, token, label in zip(sentence_list, token_list, label_list):
    line = f'{token}\t{label}'
    dev_output.append(line)
    line = (sentence, token, label)
    dev_output_hg.append(line)

with open('dev.iob', 'w', encoding='utf-8') as iob_file:
  iob_file.write('\n'.join(dev_output))
```

## Przegląd zanotowanych danych

```python
import spacy
from spacy import displacy
from spacy.tokens import DocBin
```

```python
train_doc_bin = DocBin().from_disk("./train.spacy")
dev_doc_bin = DocBin().from_disk("./dev.spacy")
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
docs = list(train_doc_bin.get_docs(nlp.vocab)) + list(dev_doc_bin.get_docs(nlp.vocab))
```

Prosty sposób na przeglądanie danych. To prosty przykład – zawsze można zmienić indeks, wyświetlić wszystkie dokumenty, opracować aplikację webową itp.:

```python
INDEX = 0
displacy.render(docs[INDEX], style="ent", jupyter=True)
```

Przykład z kolorami (można je oczywiście lepiej dobrać ;)):

```python
INDEX = 1
colors = {'Term': 'grey',
          'Definition': 'red',
          'Alias-Term': 'green',
          'Secondary-Definitions': 'violet',
          'Referential-Definition': 'blue',
          'Qualifier': 'pink',
          'Definition-frag': 'black'}
options = {'ents': ['Term', 'Definition', 'Alias-Term', 'Secondary-Definitions', 'Referential-Definition', 'Qualifier', 'Definition-frag'],
           'colors': colors}
displacy.render(docs[INDEX], style="ent", options=options, jupyter=True)
```
